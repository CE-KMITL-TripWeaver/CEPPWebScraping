```python
path_to_res_merge_accommodation_phuket = os.path.join(fh.STORE_MERGE_ACCOMM_SCRAPING, 'res_merge_accommodation', 'res_merge_accommodation_Phuket.csv')
merge_accommodation_phuket_df = pd.read_csv(path_to_res_merge_accommodation_phuket)

merge_accommodation_phuket_df = merge_accommodation_phuket_df.fillna('')
merge_accommodation_phuket_df
```

```python
def create_payload(row: pd.Series) -> dict:

    payload = {
    "name": row['name'],
    "type": ast.literal_eval(row['type']),
    "description": row['description'],
    "latitude": row['latitude'],
    "longitude": row['longitude'],
    "imgPath": ast.literal_eval(row['imgPath']),
    "phone": row['phone'],
    "website": row['website'],
    "star": row['star'],
    "facility": ast.literal_eval(row['facility']),
    "tag": ast.literal_eval(row['tag']),
    "location": {
        "address": row['address'],
        "province": row['province'],
        "province_code": row['province_code'],
        "district": row['district'],
        "district_code": row['district_code'],
        "subDistrict": row['subDistrict'],
        "sub_district_code": row['sub_district_code']
    },
    "rating": {
        "score": row['score'],
        "ratingCount": row['ratingCount']
    }
}

    return payload.copy()
```

```python
for Idx, cur_row in merge_accommodation_phuket_df.iterrows():

    cur_payload = create_payload(row=cur_row)

    try:
        API_ENDPOINT = f"{os.environ['TRIPWEAVER_API']}/api/accommodation/create"
        post_request = requests.post(url=API_ENDPOINT, data=cur_payload)
        print(post_request.text)
        
    except Exception as e:
        print("POST api failed at --> ", cur_payload['name'])
        print(e)
```
